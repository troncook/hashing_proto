```python
import pandas as pd
import numpy as np
import random
import string
import uuid
```

```python
import hashlib
```

```python
#checking out df
df = pd.read_csv('Recruiter.csv')
df.head()
```

```python
# creating a simple list of values using a list comprehension 
# based on the len (or number of rows) of the dataframe
df['uuid'] = [uuid.uuid4() for x in range(len(df))]
print(df)
```

```python
# Convert column to string
df['uuid'] = df['uuid'].astype(str)
# Apply hashing function to the column
df['EmployeeNumber_HASH'] = df['uuid'].apply(
    lambda x: 
        hashlib.sha512(x.encode()).hexdigest()
)
```

```python
#checking to see if everything looks good
df
```

```python
#saving to csv
df.to_csv('hash_combined_columns_uuid.csv', encoding='utf-8', index=False)
```
